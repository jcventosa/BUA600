```python
import pandas as pd
from matplotlib import style

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data/ufo.csv')
```

```python
data.head(2)
```

```python
#Convert datetime values from str to datetime
data['datetime-new'] = pd.to_datetime(data['datetime'],errors='coerce')
```

```python
data['datetime-new']
```

```python
data.count()
```

```python
#Drop all values with Null or NaN data
data = data.dropna(how='any')
```

```python
data.count()
```

```python
#Set DateTime column values as index
data.index = pd.to_datetime(data['datetime-new'])
```

```python
#Confirm index is correct
data
```

```python
#Set DF index to UTC Time Zone
data.index = data.index.tz_localize('utc')
```

```python
data
```

```python
#Clean up dataframe -> Remove old columns that are no longer needed
del data['datetime-new']
```

```python
data
```

```python
del data['Unnamed: 0']
```

```python
data
```

```python
data.index.names = ['date']
```

```python
data
```

```python
data['duration (seconds)'].plot(kind='line', figsize=(20,10), title='UFOs Durations across the years')
```

```python
#plotting in ggplot

style.use('ggplot')
plt.subplot(2,1,1)
plt.plot(data.index,data['duration (seconds)'])
plt.title("Duration")

plt.show()
```

```python
rolling_15_mean = data[['duration (seconds)']].rolling(15).mean()
```
